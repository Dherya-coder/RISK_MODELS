```python
import pandas as pd
import numpy as np
import cvxpy as cp
from scipy.interpolate import make_interp_spline
import matplotlib.pyplot as plt

# Read in dataset
df = pd.read_csv('CT_RATES_FINAL - Sheet1.csv')

# Fill NaN values with the mean of each column
df = df.apply(lambda col: col.fillna(col.mean()), axis=0)
col_arr = {}
arr_list = []
mu = []

# Convert data to numpy arrays and calculate mean returns
for i in range(len(df.columns)):
    col_name = df.columns[i]
    col_arr[i + 1] = df[col_name].to_numpy()
    mu.append(np.mean(col_arr[i + 1]))
    arr_list.append(col_arr[i + 1])

mu = np.array(mu)
data_matrix = np.array(arr_list)

cov_mat = np.cov(data_matrix)


n = len(mu)
T = data_matrix.shape[1]
e=np.ones(n)

l2_risk=[]
l2_return=[]
l1_risk=[]
l1_return=[]
linf_risk=[]
linf_return=[]

for i in range(100):
    w_l2=cp.Variable(n)
    lam=i/100+1/100
    objective_l2=cp.Minimize(lam*(cp.quad_form(w_l2,cov_mat))-(1-lam)*(mu.T@w_l2))
    constraints_l2 = [
    e.T @ w_l2 == 1
]
    mop_problem_l2=cp.Problem(objective_l2,constraints_l2)
    mop_problem_l2.solve()
    risk_temp_l2=np.dot(w_l2.value.T,np.dot(cov_mat,w_l2.value))
    return_temp_l2=np.dot(mu.T,w_l2.value)
    l2_risk.append(risk_temp_l2)
    l2_return.append(return_temp_l2)
    w_l1=cp.Variable(n)
    y_l1=cp.Variable(T)
    objective_l1 = cp.Minimize(lam*(1/T * cp.sum(y_l1))-(1-lam)*(mu.T@w_l1))
    deviation = data_matrix.T - mu.reshape(1, -1)
    constraints_l1=[
        deviation @ w_l1 <= y_l1,
        -deviation @ w_l1 <= y_l1,
        e.T @ w_l1 == 1,
        0<=y_l1
    ]
    mop_problem_l1=cp.Problem(objective_l1,constraints_l1)
    mop_problem_l1.solve(solver=cp.SCS,verbose=True)
    if y_l1.value is not None:
        risk_temp_l1=(1/T)*np.sum(y_l1.value)
        return_temp_l1=np.dot(mu.T,w_l1.value)
        l1_risk.append(risk_temp_l1)
        l1_return.append(return_temp_l1)
    w_linf = cp.Variable(n)
    z_linf = cp.Variable()
    constraints_linf = []
    for j in range(n):
        abs_sum = (1/T) * cp.sum([cp.abs(deviation[t, j] * w_linf[j]) for t in range(T)])
        constraints_linf.append(z_linf >= abs_sum)
    constraints_linf.append(e.T @ w_linf == 1)
    objective_linf = cp.Minimize(lam * z_linf - (1 - lam) * (mu.T @ w_linf))
    mop_problem_linf = cp.Problem(objective_linf, constraints_linf)
    mop_problem_linf.solve(solver=cp.SCS, verbose=True)
    if mop_problem_linf.status == cp.OPTIMAL:
        risk_temp_linf = z_linf.value
        return_temp_linf = np.dot(mu.T, w_linf.value)
        linf_risk.append(risk_temp_linf)
        linf_return.append(return_temp_linf)

# print(np.array(linf_risk).size)
# print(np.array(l1_risk).size)
l2_risk, l2_return = np.array(l2_risk), np.array(l2_return)
l1_risk, l1_return = np.array(l1_risk), np.array(l1_return)
linf_risk, linf_return = np.array(linf_risk), np.array(linf_return)



def plot_frontier(risk, ret, label):
    x = np.squeeze(risk)
    y = np.squeeze(ret)
    sorted_pairs = sorted(zip(x, y), key=lambda pair: pair[0])
    x, y = zip(*sorted_pairs)
    unique_x, unique_y = [], []
    for xi, yi in zip(x, y):
        if xi not in unique_x:
            unique_x.append(xi)
            unique_y.append(yi)
    unique_x = np.array(unique_x)
    unique_y = np.array(unique_y)
    X_Y_Spline = make_interp_spline(unique_x, unique_y)
    X_ = np.linspace(min(unique_x), max(unique_x), 500)
    Y_ = X_Y_Spline(X_)
    plt.plot(X_, Y_, label=label)
    plt.scatter(unique_x, unique_y, marker='o')

plot_frontier(l2_risk, l2_return, 'L2 Efficient Frontier')
plot_frontier(l1_risk, l1_return, 'L1 Efficient Frontier')
plot_frontier(linf_risk, linf_return, 'Linf Efficient Frontier')

plt.title("Efficient Frontier with L2, L1 norm and Linf norm")
plt.xlabel("variance_risk")
plt.ylabel("expected return")
plt.legend()
plt.show()
```
